### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
'''
# Check for similar column names
similar_column = 'DatabaseWantToWorkWith'
similar_columns = [col for col in df.columns if similar_column.lower() in col.lower()]
if similar_columns:
    print(f"Similar columns found: {similar_columns}")
else:
    print("No similar columns found.")
'''

# Ensure the 'DatabaseWantToWorkWith' column exists
if 'DatabaseWantToWorkWith' in df.columns:
    # Count the occurrences of each unique database
    db_counts = df['DatabaseWantToWorkWith'].value_counts()

    # Get the top 5 databases
    top_5_db = db_counts.head(5)

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(top_5_db, labels=top_5_db.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title('Top 5 Databases Respondents Wish To Work With')
    plt.show()
else:
    print("The 'DatabaseWantToWorkWith' column is not found in the dataset.")

```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Ensure the 'DevType' column exists
if 'DevType' in df.columns:
    # Count the occurrences of each unique database
    dt_counts = df['DevType'].value_counts()

    # Get the top 5 databases
    top_5_dt = dt_counts.head(5)

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(top_5_dt, labels=top_5_dt.index, autopct='%1.1f%%', startangle=90)
    plt.title('Top 5 dev types')
    plt.show()
else:
    print("The 'DevType' column is not found in the dataset.")
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
if 'OpSysProfessional' in df.columns:
    # Count the occurrences of each unique database
    dt_counts = df['OpSysProfessional'].value_counts()

    # Get the top 5 databases
    top_os = dt_counts.head(3)

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(top_os, labels=top_os.index, autopct='%1.1f%%', startangle=90)
    plt.title('Top operating systems')
    plt.show()
else:
    print("The 'OpSysProfessional' column is not found in the dataset.")
```
